```python
def check_column_types(df, col):
    # Count occurrences of each Python type in the column
    type_counts = df[col].apply(type).value_counts()
    #print(col, type_counts)

    # Return the column name if more than 1 type is present
    if len(type_counts) > 1:
        return col
    else:
        return None
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# --- 1. Original hist (100 bins) ---
def plot_hist_standard(ax, df):
    ax.hist(df["vote_count"], bins=100, edgecolor='black')
    ax.set_title("Histogram (100 bins)")
    ax.set_xlabel("vote_count")

# --- 2. Histogram with exponential bins ---
def plot_hist_exp_bins(ax, df):
    data = df["vote_count"]
    bins = [1, 100, 300, 900, 2000, 5000]

    ax.hist(data, bins=bins, edgecolor='black')
    ax.set_xscale('log')
    ax.set_yscale('log')

    ax.set_title("Histogram (exp bins)")
    ax.set_xlabel("Value")
    ax.set_ylabel("Count")

# --- 3. Cumulative frequency plot ---
def plot_cumulative(ax, df):
    votes = df["vote_count"]
    unique_votes = np.sort(np.unique(votes))
    cum_freq = np.cumsum(unique_votes)

    ax.plot(unique_votes, cum_freq, marker='o')
    ax.set_xscale('log')
    ax.set_yscale('log')

    ax.set_title("Cumulative Frequency")
    ax.set_xlabel("Votes")
    ax.set_ylabel("Cum. Frequency")

# --- 4. Zipf Plot ---
def plot_zipf(ax, df):
    votes = df['vote_count']
    sorted_votes = np.sort(votes)[::-1]
    ranks = np.arange(1, len(sorted_votes) + 1)

    ax.plot(ranks, sorted_votes, marker='o', linestyle='none')
    ax.set_xscale('log')
    ax.set_yscale('log')

    ax.set_title("Zipf Plot")
    ax.set_xlabel("Rank")
    ax.set_ylabel("Votes")

# --- 5. Seaborn histplot ---
def plot_log_votes(ax, df):
    sns.histplot(df['log_votes'], bins=100, ax=ax)
    ax.set_title("Histogram (log_votes)")



# --- Create a 2×3 grid and place the plots ---
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

plot_hist_standard(axes[0,0], df)
plot_hist_exp_bins(axes[0,1], df)
plot_cumulative(axes[0,2], df)
plot_zipf(axes[1,0], df)
plot_log_votes(axes[1,1], df)

# Hide the unused last subplot (bottom right)
axes[1,2].axis('off')

plt.tight_layout()
plt.show()
```

t-tests

```python
import scipy.stats as stats

group1 = final_df[final_df["genre_name"] == "Action" ]["revenue"].dropna()
group2 = final_df[final_df["genre_name"] == "Drama"]["revenue"].dropna()

t, p = stats.ttest_ind(group1, group2, equal_var=False)
print("p-value:", p)
print("t-value:",t)
```

```python
import scipy.stats as stats

group1 = final_df[final_df["genre_name"] == "Action"]["budget"].dropna()
group2 = final_df[final_df["genre_name"] == "Drama"]["budget"].dropna()

t, p = stats.ttest_ind(group1, group2, equal_var=False)
print("p-value:", p)
print("t-value:",t)

```

```python
import scipy.stats as stats

group1 = final_df[final_df["genre_name"] == "Action"]["vote_average"].dropna()
group2 = final_df[final_df["genre_name"] == "Drama"]["vote_average"].dropna()

t, p = stats.ttest_ind(group1, group2, equal_var=False)
print("p-value:", p)
print("t-value:",t)
```

Από τα στατιστικά t-test, παρατηρείται πως και στα τρία αποτελέσματα η τιμή του pvalue είναι μικρή, άρα μεγάλης σημαντικότητας. Επίσης και στα τρία τεστ η τιμή t είναι αρκετά μεγάλη, κάτι που σημαίνει πως υπάρχει διαφορά μεταξύ των δεδομένων revenue, budget, vote_average σχετικά με το φύλο του σκηνοθέτη, με τη μεγαλύτερη διαφορά στο budget που έχει ο σκηνοθέτης του κάθε φύλου (|t|~|14.4|).
