# UA and IA (step 1)

```python
df_train = df_train.loc[(df_train!=0).any(axis=1)] #did nothing, wanted to eliminate all 0 ratings- non existent ratings
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import root_mean_squared_error

# train and test dataframes assumed to have columns: userId, movieId, rating

# === 1. Compute user averages ===
user_mean = df_train.groupby("uid")["rating"].mean()

# === 2. Compute movie averages ===
movie_mean = df_train.groupby("mid")["rating"].mean()

# === 3. Global mean (backup if user/movie not in train set)
global_mean = df_train["rating"].mean()

# ---------------------------------------------
# User Average Predictions
# ---------------------------------------------
ua_pred = df_test["uid"].map(user_mean)    # maps user to mean rating
ua_pred = ua_pred.fillna(global_mean)      # in case user not in train

# RMSE UA
rmse_ua = root_mean_squared_error(df_test["rating"], ua_pred)


# ---------------------------------------------
# Item Average Predictions
# ---------------------------------------------
ia_pred = df_test["mid"].map(movie_mean)  # maps movie to mean rating
ia_pred = ia_pred.fillna(global_mean)      # in case movie not in train

# RMSE IA
rmse_ia = root_mean_squared_error(df_test["rating"], ia_pred)


print("RMSE User Average:", rmse_ua)
print("RMSE Item Average:", rmse_ia)
```
